## Import libraries

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1)
sample_size = 1000
```

## Defining a hidden relationship (4 variables)

```python
x1 = np.round(np.random.normal(0,200,sample_size),2)
x2 = np.round(np.random.normal(10,35,sample_size),2)
x3 = np.round(np.random.normal(100,200,sample_size),2)
x4 = np.round(np.random.normal(30,30,sample_size),2)

# Define a linear model with the following parameter values


b0 = 1
b1 = -50.5
b2 = 10.5
b3 = 100.5
b4 = -520.45


# The hidden relationship between the variables is as follows

y = b0 + b1 * x1 + b2 *x2 + b3 * x3 + b4 * x4
```

## Adding noise to the model

```python
e = np.round(np.random.normal(0,50,sample_size),2)

Y = y + e
```

```python
fig, ax = plt.subplots(2,2,figsize=(20,15))

ax[0,0].scatter(x = x1, y = y)
ax[0,0].set_xlabel('x1 Input')
ax[0,0].set_ylabel('Target Values')

ax[0,1].scatter(x = x2, y = y)
ax[0,1].set_xlabel('x2 Input')
ax[0,1].set_ylabel('Target Values')

ax[1,0].scatter(x = x3, y = y)
ax[1,0].set_xlabel('x3 Input')
ax[1,0].set_ylabel('Target Values')

ax[1,1].scatter(x = x4, y = y)
ax[1,1].set_xlabel('x4 Input')
ax[1,1].set_ylabel('Target Values')

plt.show()
```

## Create a dataframe

```python
import pandas as pd

df = pd.DataFrame({'x1' : x1, 'x2' : x2, 'x3' : x3, 'x4' : x4, 'target' : Y})

df.head()
```

```python
df.to_csv('dataset_multivariable_regression.csv', index = False)
```
