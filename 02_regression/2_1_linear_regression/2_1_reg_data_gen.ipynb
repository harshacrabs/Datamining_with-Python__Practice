```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Set random seed

```python
np.random.seed(1)
```

```python
b1 = 30.5 #slope
b0 = 20 #y-intercept
```

## Generate data

```python
sample_size = 1000

x_mean = 100
x_stdev = 10
X = np.round(np.random.normal(x_mean, x_stdev, sample_size),2)
```

```python
y = b0 + b1 * X
```

```python
plt.scatter(X,y)
plt.show()
```

## Generate noise

```python
e_mean = 0
e_stdev = 100
e = np.round(np.random.normal(e_mean, e_stdev, sample_size),2)

Y = y+e
```

```python
plt.scatter(X,Y)

plt.show()
```

```python
df = pd.DataFrame({'input': X, 'target' : Y})

df
```

```python
df.to_csv('reg_synthetic_dataset.csv', index=False)
```
